```python
import requests
import time

def monitor_recent_draws():
    API_URL = "https://api.merkle.trade/v1/season/s29/dashboard/recent-draws"  # Fill this in with the actual API URL
    THRESHOLD = 0.75  # Minimum reward amount to trigger a "big win" alert

    # Initialize a persistent set for processed draws to avoid duplicates
    processed_draws = set()

    try:
        response = requests.get(API_URL)
        response.raise_for_status()  # Raise error for bad responses
        data = response.json()

        for draw in data["items"]:
            user = draw.get("user", "Unknown User")
            reward_amount = draw["reward"]["amount"]
            reward_price = draw["reward"]["price"]
            timestamp = draw.get("drawAt", "No Timestamp")

            # Generate a unique identifier (draw_id) based on the user and timestamp
            draw_id = f"{user}-{timestamp}-{reward_amount}-{reward_price}"

            # Check if the draw has already been processed
            if draw_id in processed_draws:
                continue

            # Add this draw to the processed set
            processed_draws.add(draw_id)

            # Check if the reward exceeds the threshold
            if reward_amount >= THRESHOLD:
                message = f"Big Win Alert! User: {user}, Amount: {reward_amount} APT, Timestamp: {timestamp}, Price: ${reward_price}"

                # Send the message via Telegram
                token = "7674390801:AAE29eHavtDHf-4fP1Jcoe_ugElTgJbx6tU"  # Your bot token
                chat_id = "6012135285"  # Your actual chat ID
                url = f"https://api.telegram.org/bot{token}/sendMessage"
                params = {'chat_id': chat_id, 'text': message}
                response = requests.get(url, params=params)

                # Check if the message was sent successfully
                if response.status_code == 200:
                    print(f"Message sent successfully for User: {user}")
                else:
                    print(f"Failed to send message for User: {user}, Status Code: {response.status_code}")

                print("Waiting for next cycle...")

            time.sleep(10)  # Check every 10 seconds

    except Exception as e:
        print(f"Error: {e}")
        time.sleep(10)  # Wait before checking again in case of errors
```

```python
# Start monitoring for big wins...
while True:
    monitor_recent_draws()  # Monitor function call
    print("Waiting for next cycle...")  # For logging purposes in Colab
    time.sleep(10)  # Check every 10 seconds
```
